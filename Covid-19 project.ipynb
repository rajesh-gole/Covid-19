```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
web_page = pd.read_html("https://en.wikipedia.org/wiki/COVID-19_pandemic_in_West_Bengal","Recoveries")[1]
```

```python
rec = pd.Series(map(int,web_page['recoveries']))
TC = pd.Series(map(int,web_page['total cases']))
death = pd.Series(map(int,web_page['deaths']))
AC = pd.Series(map(int,web_page['active cases']))
```

```python
web_page.columns = ['states','total cases','recoveries','deaths','active cases']
```

```python
web_page.drop([24],inplace=True)
```

```python
web_page
```

```python
fig = plt.figure(figsize=(15,10))
axes = fig.add_axes([0,0,1,1])
x = np.arange(0,4*24,step=4)
axes.bar(x,TC,width = 2, label="Total Cases")
axes.bar(x+2,rec,width = 2,label="Recoveries")
axes.bar(x+4,AC,width = 2,label="Active Cases",color="yellow")
axes.bar(x+6,death,width = 2,label="Deaths",color="c")
plt.xticks(x+2,web_page["states"],rotation=80)
axes.legend()
plt.show()
```
